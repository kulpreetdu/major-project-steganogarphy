# histogram of carrier image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('lady.jpg',0)
#img.ravel()- converting the image into flattened one dimensional array
plt.hist(img.ravel(),256,[0,256]); plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('protest.png',0)
#img.ravel()- converting the image into flattened one dimensional array
plt.hist(img.ravel(),256,[0,256]); plt.show()
```

# histogram in rgb form of carrier image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('lady.jpg')
color = ('r','g','b')
for i,col in enumerate(color):
    print(i, col)
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
